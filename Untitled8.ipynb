Now, let's import the necessary libraries and define our input and output paths. We'll also create the output directory if it doesn't already exist.

```python
import os
from PIL import Image

# Define input and output directories
input_base_dir = '/content/drive/MyDrive/cricket'
output_base_dir = '/content/resized_cricket_images'

# Desired image size
output_size = (250, 250)

# Create the output base directory if it doesn't exist
os.makedirs(output_base_dir, exist_ok=True)

print(f"Input directory: {input_base_dir}")
print(f"Output directory: {output_base_dir}")
```

Finally, we will loop through the 'cricket Ball' and 'cricket Bat' subfolders, resize each image, and save it to the corresponding subfolder within the new output directory.

```python
for subfolder_name in os.listdir(input_base_dir):
    subfolder_path = os.path.join(input_base_dir, subfolder_name)
    output_subfolder_path = os.path.join(output_base_dir, subfolder_name)

    # Ensure it's a directory and create corresponding output subfolder
    if os.path.isdir(subfolder_path):
        os.makedirs(output_subfolder_path, exist_ok=True)
        print(f"Processing images in: {subfolder_path}")

        for filename in os.listdir(subfolder_path):
            if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
                input_image_path = os.path.join(subfolder_path, filename)
                output_image_path = os.path.join(output_subfolder_path, filename)

                try:
                    with Image.open(input_image_path) as img:
                        img_resized = img.resize(output_size, Image.LANCZOS)
                        img_resized.save(output_image_path)
                        # print(f"Resized and saved: {filename}")
                except Exception as e:
                    print(f"Error processing {filename}: {e}")
    else:
        print(f"Skipping non-directory item: {subfolder_path}")

print("\nImage resizing and saving complete!")
print(f"Resized images are saved in: {output_base_dir}")
```

Now, let's perform image augmentation on these resized images. We'll use `ImageDataGenerator` from Keras to apply various transformations and save the augmented images to a new directory.

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np

# Define the input directory (where the resized images are) and output directory for augmented images
input_resized_dir = '/content/resized_cricket_images'
output_augmented_dir = '/content/augmented_cricket_images'

# Create the output base directory for augmented images if it doesn't exist
os.makedirs(output_augmented_dir, exist_ok=True)

# Define augmentation parameters
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

print(f"Input directory for augmentation: {input_resized_dir}")
print(f"Output directory for augmented images: {output_augmented_dir}")
```

Next, we will loop through the subfolders in the `resized_cricket_images` directory, apply the augmentation, and save a few augmented versions of each image to the `augmented_cricket_images` directory.

```python
num_augmented_images_per_original = 5 # Number of augmented versions to create for each original image

for subfolder_name in os.listdir(input_resized_dir):
    subfolder_path = os.path.join(input_resized_dir, subfolder_name)
    output_aug_subfolder_path = os.path.join(output_augmented_dir, subfolder_name)

    if os.path.isdir(subfolder_path):
        os.makedirs(output_aug_subfolder_path, exist_ok=True)
        print(f"Augmenting images in: {subfolder_path}")

        for filename in os.listdir(subfolder_path):
            if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
                img_path = os.path.join(subfolder_path, filename)
                img = Image.open(img_path)
                img_array = np.array(img) # Convert PIL image to numpy array
                img_array = img_array.reshape((1,) + img_array.shape) # Reshape to (1, height, width, channels)

                i = 0
                for batch in datagen.flow(img_array, batch_size=1,
                                          save_to_dir=output_aug_subfolder_path,
                                          save_prefix='aug_',
                                          save_format='jpeg'): # Save as JPEG for consistency
                    i += 1
                    if i >= num_augmented_images_per_original:
                        break # Stop after generating 'num_augmented_images_per_original' augmented images
    else:
        print(f"Skipping non-directory item: {subfolder_path}")

print("\nImage augmentation complete!")
print(f"Augmented images are saved in: {output_augmented_dir}")
```

Now that we have augmented images, let's prepare them for training a Convolutional Neural Network (CNN). We'll use `ImageDataGenerator` again to create data generators for our training and validation sets, and then define and train a simple CNN model.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os

# Define the directory containing the augmented images
data_dir = '/content/augmented_cricket_images'

# Define image dimensions and batch size
img_height, img_width = 250, 250 # Should match the size used for resizing
batch_size = 32

# Define the actual classes you want to classify
target_classes = ['cricket ball', 'cricket bat']

# Create an ImageDataGenerator for loading data into the CNN
# We'll use a small validation split to test the model setup
train_datagen = ImageDataGenerator(
    rescale=1./255, # Normalize pixel values to [0, 1]
    validation_split=0.2 # 20% of data will be used for validation
)

# Create training and validation data generators
train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary', # Since we have two classes (cricket Ball, cricket Bat)
    classes=target_classes, # Explicitly define classes to exclude .ipynb_checkpoints
    subset='training',
    seed=42 # for reproducibility
)

validation_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary',
    classes=target_classes, # Explicitly define classes to exclude .ipynb_checkpoints
    subset='validation',
    seed=42
)

print(f"Found {train_generator.num_classes} classes: {train_generator.class_indices}")
print(f"Number of training samples: {train_generator.samples}")
print(f"Number of validation samples: {validation_generator.samples}")
```

Now, let's define a simple Convolutional Neural Network (CNN) architecture. This model will consist of convolutional layers, pooling layers, and dense layers for classification.

```python
# Build the CNN model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(img_height, img_width, 3)),
    MaxPooling2D((2, 2)),

    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),

    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5), # Add dropout for regularization
    Dense(1, activation='sigmoid') # Sigmoid for binary classification
])

# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Display model summary
model.summary()
```

Finally, let's train the CNN model using the `train_generator` and `validation_generator`.

```python
# Train the model
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    epochs=10, # You can adjust the number of epochs
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // batch_size
)

print("\nModel training complete!")
```

### `app.py` for Streamlit Deployment

Save the following code as `app.py` in the same directory as your `cricket_classifier_model.h5` file. This script will create a simple web application where you can upload an image and get a prediction from your trained model.

```python
%%writefile app.py

import streamlit as st
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np
import os

# Load the trained model
# Using suppress_warnings=True because we saved it as .h5 which Keras warns about
@st.cache_resource
def load_model():
    model = tf.keras.models.load_model('cricket_classifier_model.h5', compile=False)
    # Recompile the model after loading, if necessary (e.g., if you plan to continue training)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

model = load_model()

# Define image dimensions and classes (must match training)
IMG_HEIGHT, IMG_WIDTH = 250, 250
CLASS_NAMES = ['cricket ball', 'cricket bat']

st.title('Cricket Ball vs. Cricket Bat Classifier')
st.write('Upload an image of a cricket item to classify it as a ball or a bat.')

# Image Upload
uploaded_file = st.file_uploader("Choose an image...", type=["jpg", "jpeg", "png"])

if uploaded_file is not None:
    # Display the uploaded image
    st.image(uploaded_file, caption='Uploaded Image.', use_column_width=True)
    st.write("")
    st.write("Classifying...")

    try:
        # Preprocess the image
        img = image.load_img(uploaded_file, target_size=(IMG_HEIGHT, IMG_WIDTH))
        img_array = image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0) # Add batch dimension
        img_array /= 255.0 # Normalize pixel values

        # Make prediction
        predictions = model.predict(img_array)

        # Interpret prediction for binary classification
        if predictions[0][0] > 0.5:
            predicted_class_name = CLASS_NAMES[1] # 'cricket bat'
            confidence = predictions[0][0]
        else:
            predicted_class_name = CLASS_NAMES[0] # 'cricket ball'
            confidence = 1 - predictions[0][0]

        st.success(f"Prediction: **{predicted_class_name}**")
        st.info(f"Confidence: {confidence:.2f}")

    except Exception as e:
        st.error(f"Error processing image: {e}")
        st.warning("Please ensure the uploaded image is valid and try again.")
```

Now, let's create a function to perform a single image prediction using our trained model.

```python
from tensorflow.keras.preprocessing import image
import numpy as np

def predict_single_image(model, img_path, target_size, class_indices):
    img = image.load_img(img_path, target_size=target_size)
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  # Create a batch
    img_array /= 255.0  # Normalize the image pixels

    prediction = model.predict(img_array)

    # Interpret the prediction
    if class_indices is None:
        print("Class indices not provided. Cannot map prediction to class name.")
        return prediction

    # Reverse the class_indices dictionary to map integer to class name
    idx_to_class = {v: k for k, v in class_indices.items()}

    if model.output_shape[-1] == 1: # Binary classification with sigmoid
        if prediction[0][0] > 0.5:
            predicted_class_idx = 1
            confidence = prediction[0][0]
        else:
            predicted_class_idx = 0
            confidence = 1 - prediction[0][0]
        predicted_class_name = idx_to_class[predicted_class_idx]
        print(f"Prediction: {predicted_class_name} (Confidence: {confidence:.2f})")
    else: # Multi-class classification with softmax
        predicted_class_idx = np.argmax(prediction[0])
        predicted_class_name = idx_to_class[predicted_class_idx]
        confidence = prediction[0][predicted_class_idx]
        print(f"Prediction: {predicted_class_name} (Confidence: {confidence:.2f})")

    return predicted_class_name, confidence

# Example usage (replace with an actual image path from your augmented data)
sample_image_path = os.path.join(output_augmented_dir, 'cricket bat', 'aug_0_10bat.jpeg') # Adjust this path as needed

# Ensure the target_classes and class_indices from train_generator are available
# If you restart the kernel, you might need to re-run the ImageDataGenerator cell
if 'train_generator' in locals():
    class_indices_map = train_generator.class_indices
else:
    print("train_generator not found. Please ensure the data generator cell has been run.")
    class_indices_map = {'cricket ball': 0, 'cricket bat': 1} # Fallback, assumes these indices


if os.path.exists(sample_image_path):
    print(f"Predicting for image: {sample_image_path}")
    predicted_class, confidence = predict_single_image(model, sample_image_path, (img_height, img_width), class_indices_map)
else:
    print(f"Sample image not found at {sample_image_path}. Please provide a valid path to an image for prediction.")
```

### Visualizing Augmented Images

Let's display a few augmented images from each class to see the transformations applied and confirm the data quality.

```python
import matplotlib.pyplot as plt
import random

def display_random_augmented_images(data_dir, num_images_per_class=3):
    classes = [d for d in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, d)) and not d.startswith('.')]

    plt.figure(figsize=(15, 8))
    plt.suptitle('Sample Augmented Images', fontsize=16)

    img_count = 0
    for class_name in classes:
        class_path = os.path.join(data_dir, class_name)
        image_files = [f for f in os.listdir(class_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

        if not image_files:
            print(f"No image files found in {class_path}")
            continue

        # Select random images from the current class
        selected_images = random.sample(image_files, min(num_images_per_class, len(image_files)))

        for img_filename in selected_images:
            img_count += 1
            img_path = os.path.join(class_path, img_filename)
            img = Image.open(img_path)

            plt.subplot(len(classes), num_images_per_class, img_count)
            plt.imshow(img)
            plt.title(f"{class_name}\n({img_filename[:20]}...)") # Show first 20 chars of filename
            plt.axis('off')

    plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent title overlap
    plt.show()

# Display 3 random augmented images from each class
display_random_augmented_images(output_augmented_dir, num_images_per_class=3)
```
